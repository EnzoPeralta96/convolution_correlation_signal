# Experimentos de Detección de Patrones en Audio

Sistema de detección de patrones de reggaeton usando correlación cruzada normalizada.

## Configuración
- **Patrón:** Ritmo característico de reggaeton
- **Umbral (τ):** 0.01
- **Sample Rate:** 44100 Hz

## Matriz de Confusión
- **VP (Verdaderos Positivos):** Canciones reggaeton correctamente detectadas
- **VN (Verdaderos Negativos):** Canciones no-reggaeton correctamente rechazadas
- **FP (Falsos Positivos):** Canciones no-reggaeton incorrectamente detectadas como reggaeton
- **FN (Falsos Negativos):** Canciones reggaeton incorrectamente rechazadas

```python
import os
import librosa
from lib import (
    cargar_audio_y_patron,
    correlacion_rapida_fft,
    plot_espectros,
    plot_canciones,
    plot_resultado,
    evaluar_deteccion,
    imprimir_resultado
)
```

```python
# Configuración global
TASA_MUESTREO = 44100  # Tasa de muestreo estándar (44.1kHz)
TAU = 0.01  # Umbral para la correlación
PATRON_FILE = os.path.join(os.getcwd(), 'data', 'patron.wav')
```

```python
# Cargar patrón una sola vez (se reutilizará en todos los experimentos)
print("Cargando patrón de reggaeton...")
p, _ = librosa.load(PATRON_FILE, sr=TASA_MUESTREO)
print(f"Patrón cargado: {len(p)} muestras\n")
```

## 1. Verdaderos Negativos (VN)

Canciones **no-reggaeton** que el sistema debe **rechazar correctamente**.

**Objetivo:** R_max ≤ τ (0.01) → Sistema debe clasificar como "NO reggaeton"

```python
print("=== VN: Canción No-Reggae 1 ===\n")
CANCION_FILE = os.path.join(os.getcwd(), 'data', 'cancion_no_reggae_1.wav')

# Cargar solo la canción (el patrón ya está cargado como 'p')
x, _ = librosa.load(CANCION_FILE, sr=TASA_MUESTREO)
print(f"Canción cargada: {len(x)} muestras\n")

r, lags = correlacion_rapida_fft(x, p)

plot_canciones(x, p, TASA_MUESTREO)
plot_espectros(x, p, TASA_MUESTREO)
R_max, t_peak = plot_resultado(r, lags, TAU, TASA_MUESTREO)

resultado = evaluar_deteccion(R_max, TAU, es_positivo_esperado=False)
imprimir_resultado(resultado)
```
